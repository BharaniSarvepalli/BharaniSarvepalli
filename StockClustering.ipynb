```python
data_frame.fillna(0,inplace = True)
data_frame.replace(0,inplace = True)
# Define a stock_open array which is the transpose of data_frame['Open']
stock_open = np.array(data_frame['Open']).T
stock_high = np.array(data_frame['High']).T
stock_low = np.array(data_frame['Low']).T
# Define a stock_close array which is the transpose of data_frame['Close']
stock_close = np.array(data_frame['Close']).T


#print(stock_close.iloc[0])
row, col = stock_close.shape
# create movements dataset filled with 0's
movements = np.zeros([row, col])
for i in range(0, row):
    movements[i,:] = np.subtract(stock_close[i,:], stock_open[i,:])
# Display the sums of the movement for each company defined above
#for i in range(len(dict_companies)):
#    print('Company: {}, Change: {}'.format(dict_companies[i][0], sum_movements[i][0]))
```

```python
# Normalize the movement
from sklearn.preprocessing import Normalizer
normalizer = Normalizer()

# Fit and Transform our movement data
normalized_movement = normalizer.fit_transform(movements)

# Display the minimum, maximum and mean of the normalized movement data
print('Minimim: ', normalized_movement.min())
print('Maximim: ', normalized_movement.max())
print('Mean: ', normalized_movement.mean())
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.cluster import KMeans

# Initialize a new normalizer
normalizer2 = Normalizer()

# Define a KMeans model
kmeans_model = KMeans(n_clusters = 4, max_iter = 1000, random_state = 1)

# Define a pipeline which combines our normalizer and KMeans model
pipeline = make_pipeline(normalizer2, kmeans_model)

# Fit our model to the un-normalized movement data
pipeline.fit(movements)

# Generate predictins of the clusters based on the un-normalized movement data
predictions = pipeline.predict(movements)
```

```python
# Reduce to 2 dimensions

from sklearn.decomposition import PCA

# Initialize normalizer
normalizer3 = Normalizer()

# Reduce the dimensions to 2
pca_data = PCA(n_components = 2)

# Define a new KMeans model
kmeans_pca = KMeans(n_clusters = 3, max_iter = 1000)

# Define a new pipeline combining our normalizer, pca and KMeans model
pipeline_pca = make_pipeline(normalizer3, pca_data, kmeans_pca)

# Fit our new pipeline to un-normalized movement data
pipeline_pca.fit(movements)

# Generate predictins of the clusters based on the un-normalized movement data
predictions_pca = pipeline_pca.predict(movements)

# Define a data frame to hold companies and their clusters
data_frame_clusters_pca = pd.DataFrame({
    'Cluster' : predictions_pca,
    'Comapnies' : list(dict_companies)
  #  'Center': kmeans_pca.cluster_centers_
})
```

```python
# Reduce the data
pca_data = PCA(n_components = 2).fit_transform(normalized_movement)

# Define the step size of mesh
h = 0.01

# Plot the decision boundary using a mesh grid
x_min, x_max = pca_data[:,0].min() - 1, pca_data[:,0].max() + 1
y_min, y_max = pca_data[:,1].min() - 1, pca_data[:,1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

# Obtain labels for each point in the mesh grid using our trained model
kpredictions = kmeans_pca.predict(np.c_[xx.ravel(), yy.ravel()])
kpredictions = kpredictions.reshape(xx.shape)

# Color Plot
cmap = plt.cm.Paired

# Plot the figure
plt.clf()
plt.figure(figsize = (10, 10))
plt.imshow(kpredictions, interpolation = 'nearest', extent = (xx.min(), xx.max(), yy.min(), yy.max()), cmap = cmap, aspect = 'auto', origin = 'lower')
plt.plot(pca_data[:,0], pca_data[:,1], 'k.', markersize = 5)

# Plot the centroid of each cluster
centroids = kmeans_pca.cluster_centers_
plt.scatter(centroids[:,0], centroids[:,1], marker = 'x', s = 169, linewidths = 3, color = 'w', zorder = 10)
plt.title('Stock Market Clustering Using K-Means (PCA-Reduced Data)')
plt.xlim(x_min, x_max)
plt.xlim(y_min, y_max)

plt.show()
```
